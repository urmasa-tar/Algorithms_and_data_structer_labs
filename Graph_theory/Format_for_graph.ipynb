```python
import numpy as np
import pandas as pd
```

```python
grap_part = pd.read_excel('points_of_map.xlsx', index_col=0)

grap_part
```

```python
arr_of_ind = list(grap_part.index.values)  
```

```python
for col in grap_part.columns:
    for index in arr_of_ind:
        if( index == col):
            continue
        elif(index > col):
            mir_col = index
            mir_ind = col
            if np.isnan(grap_part.loc[col, index]):
                continue
            else:
                grap_part.loc[mir_col, mir_ind] = grap_part.loc[col, index]
        else:
            continue
```

```python
print(grap_part)

grap_part.to_csv('result_matrix.csv', index=False)
```

```python
lst = []

for col in grap_part.columns:
    for index in arr_of_ind:
        if(index == col):
            continue
        else:
            if np.isnan(grap_part.loc[col, index]):
                continue
            else:
                local_fr = col
                local_to = index
                ln = grap_part.loc[col, index]
                df2 = {names_of_col[0]: local_fr, names_of_col[1]: local_to, names_of_col[2]: ln} 
                lst.append(df2)


second_teble = pd.DataFrame(lst, columns=['from', 'to', 'length'])

print(second_teble)

second_teble.to_csv('res_row_mat.csv', index=False)
```
